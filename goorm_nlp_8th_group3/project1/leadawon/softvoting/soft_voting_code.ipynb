```python
# 모델별 예측값 불러오기

csv1 = pd.read_csv('/content/bert_base_uncased_softvoting.csv')
csv2 = pd.read_csv('/content/distilbert_base_uncased_softvoting.csv')
csv3 = pd.read_csv('/content/roberta_base_uncased_softvoting.csv')
#csv4 = pd.read_csv('')
#csv5 = pd.read_csv('')
#csv6 = pd.read_csv('')
```

```python
ex0_1 = csv1['ex0']
ex0_2 = csv2['ex0']
ex0_3 = csv3['ex0']
#ex0_4 = csv4['ex0']
#ex0_5 = csv5['ex0']
#ex0_6 = csv6['ex0']
```

```python
ex1_1 = csv1['ex1']
ex1_2 = csv2['ex1']
ex1_3 = csv3['ex1']
#ex1_4 = csv4['ex1']
#ex1_5 = csv5['ex1']
#ex1_6 = csv6['ex1']
```

```python
ex0_df = pd.DataFrame({'ex0_1':ex0_1, 'ex0_2':ex0_2, 'ex0_3':ex0_3})
```

```python
ex1_df = pd.DataFrame({'ex1_1':ex1_1, 'ex1_2':ex1_2, 'ex1_3':ex1_3})
```

```python
ex0_df
```

```python
# 모델들이 예측한 0, 1 중 더 많이 나온 것 선택(voting) 
# df.loc['Sum'] = df.apply(np.sum, axis = 0)
def find_most_frequent(row):
    suma = row.sum()
    return suma

# Apply the function to the DataFrame and store the result in a new column
import numpy as np
pred_df = pd.DataFrame()
pred_df['ex0'] = ex0_df.apply(find_most_frequent, axis=1)
pred_df['ex1'] = ex1_df.apply(find_most_frequent, axis=1)

print(pred_df)
```

```python
#df['add'] = df.apply(lambda row : add(row['A'], row['B'], row['C']), axis = 1)
pred_df['most_frequent'] = pred_df.apply(lambda row : 0 if row['ex0']>row['ex1'] else 1,axis=1)
```
