```python
# 모델별 예측값 불러오기

pred1 = pd.read_csv('submission_albert_base_v2.csv')
pred2 = pd.read_csv('submission_bert_base_uncased_matplotlib.csv')
pred3 = pd.read_csv('submission_bert_large_uncased.csv')
pred4 = pd.read_csv('submission_distilbert.csv')
pred5 = pd.read_csv('submission_roberta_base_uncased.csv')
pred6 = pd.read_csv('submission_roberta_yelp_finetuned.csv')
```

```python
pred1 = pred1['Category']
pred2 = pred2['Category']
pred3 = pred3['Category']
pred4 = pred4['Category']
pred5 = pred5['Category']
pred6 = pred6['Category']
```

```python
pred_df = pd.DataFrame({'pred1':pred1, 'pred2':pred2, 'pred3':pred3,'pred4':pred4,'pred5':pred5,'pred6':pred6,})
```

```python
pred_df
```

```python
# 모델들이 예측한 0, 1 중 더 많이 나온 것 선택(voting) 

def find_most_frequent(row):
    counts = row.value_counts()
    return counts.idxmax()

# Apply the function to the DataFrame and store the result in a new column
pred_df['most_frequent'] = pred_df.apply(find_most_frequent, axis=1)

print(pred_df)
```

```python
test_df = pd.read_csv('submission_albert_base_v2.csv')
test_df['Category'] = pred_df['most_frequent']
```

```python
test_df.to_csv('submission_vote_old.csv', index=False)
```
